```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Charger les données
X_train = df.drop(['Date/Time','target_vent','target_hum','target_tmp'], axis=1).iloc[:8759]  # Premières 8771 lignes pour l'entraînement
y_train = df['target_tmp'].iloc[:8759]

X_test = df.drop(['Date/Time','target_vent','target_hum','target_tmp'], axis=1).iloc[-24:]  # Dernières 12 lignes pour le test
y_test = df['target_tmp'].iloc[-24:]
# Choisissez le modèle de régression (dans cet exemple, une régression linéaire)
model = LinearRegression()

# Entraînez le modèle sur les données d'entraînement
model.fit(X_train, y_train)

# Faites des prédictions sur les données de test
predictions = model.predict(X_test)

# Évaluez les performances du modèle
mse = mean_squared_error(y_test, predictions)
print('Mean Squared Error:', mse)
```

## predict all day all hours

```python
import pandas as pd
import numpy as np

# Supposons que new_data est votre dataframe d'origine
new_data = pd.DataFrame({
    'Date/Time': ['00:00'],
    'Temp_C': [last_temp['Température Réelle']],
    'Rel Hum_%': [last_temp['Humidity']],
    'Wind Speed_km/h': [last_wind['vent speed']]
})

# Nombre d'heures à prédire
hours_to_predict = 23

# Créer un dataframe pour stocker toutes les prédictions
all_predictions = pd.DataFrame(columns=['Date/Time', 'Temp_C', 'Rel Hum_%', 'Wind Speed_km/h'])
all_predictions = pd.concat([all_predictions, new_data], ignore_index=True)

# Copier les données d'origine pour le processus de prédiction
data = new_data.copy()

for hour in range(1, hours_to_predict + 1):  
    # Sélectionner les données pour les 24 dernières heures
    data_new = data.drop(['Date/Time'], axis=1)

    # Prédire les valeurs pour la prochaine heure
    prediction_vent = model2.predict(data_new)
    prediction_hum = model1.predict(data_new)
    prediction_tmp = model.predict(data_new)
    
    # Stocker les prédictions dans un dataframe temporaire
    temp_df = pd.DataFrame({
        'Date/Time': [f'{hour:02d}:00'],
        'Temp_C': prediction_tmp,
        'Rel Hum_%': prediction_hum,
        'Wind Speed_km/h': prediction_vent
    })
    
    # Ajouter la nouvelle ligne à all_predictions
    all_predictions = pd.concat([all_predictions, temp_df], ignore_index=True)

    # Mettre à jour les données d'entrée pour la prochaine prédiction
    data = pd.concat([data, temp_df], ignore_index=True)
    data = data.iloc[1:]

# Afficher toutes les prédictions
print(all_predictions)
```

```python
from pymongo import MongoClient
import pandas as pd

# Connexion à MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["station_meteorologique"]
collection = db["predict all hours"]

# Itérer sur le DataFrame et insérer chaque ligne dans la collection MongoDB
for index, row in all_predictions.iterrows():
    prediction_doc = {
        "Date/Time": row["Date/Time"],
        "Temp_C": row["Temp_C"],
        "Rel Hum_%": row["Rel Hum_%"],
        "Wind Speed_km/h": row["Wind Speed_km/h"],
    }

    # Insérer le document de prédiction dans la collection MongoDB
    insert_result = collection.insert_one(prediction_doc)
    
    # Vérifier le résultat de l'insertion
    if insert_result.acknowledged:
       print(f"Prédiction pour Date/Time {row['Date/Time']} insérée avec succès dans MongoDB.")
    else:
       print(f"Erreur lors de l'insertion de la prédiction pour Date/Time {row['Date/Time']} dans MongoDB.")
       print("Détails de l'erreur:", insert_result.raw_result)  # Afficher les détails de l'erreur retournés par MongoDB
```

```python
# Calculer la moyenne de l'humidité et de la vitesse du vent par jour
daily_avg_humidity = df.groupby('Date')['Rel Hum_%'].mean().reset_index()
daily_avg_wind_speed = df.groupby('Date')['Wind Speed_km/h'].mean().reset_index()

# Renommer les colonnes
daily_avg_temp=daily_avg_temp.rename(columns={'Temp_C':'Moyenne de température'})
daily_avg_humidity = daily_avg_humidity.rename(columns={'Rel Hum_%': 'Moyenne Humidité Relative (%)'})
daily_avg_wind_speed = daily_avg_wind_speed.rename(columns={'Wind Speed_km/h': 'Moyenne Vitesse du Vent (km/h)'})

# Arrondir les valeurs moyennes et les convertir en entiers
daily_avg_humidity['Moyenne Humidité Relative (%)'] = daily_avg_humidity['Moyenne Humidité Relative (%)'].round().astype(int)
daily_avg_wind_speed['Moyenne Vitesse du Vent (km/h)'] = daily_avg_wind_speed['Moyenne Vitesse du Vent (km/h)'].round().astype(int)

# Fusionner les DataFrames
daily_avg_temp = pd.merge(daily_avg_temp, daily_avg_humidity, on='Date')
daily_avg_temp = pd.merge(daily_avg_temp, daily_avg_wind_speed, on='Date')

# Afficher le DataFrame final
print(daily_avg_temp)
```
