```python
import pandas as pd


weather = pd.read_csv("../Data/Weather_Data.csv")
```

```python
df=weather
```

```python
# Convertir la colonne "Date/Time" en format de date/heure
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%m/%d/%Y %H:%M', errors='coerce')

# Fonction pour remplacer l'année
def replace_year(date):
    if pd.notnull(date) and date.year == 2012:
        try:
            return date.replace(year=2023)
        except ValueError:
            pass
    return date

# Appliquer la fonction à la colonne "Date/Time"
df['Date/Time'] = df['Date/Time'].apply(replace_year)

# Afficher les dix premières lignes pour vérification
print(df.head(8000))
```

```python
df.drop(columns=["Dew Point Temp_C"], inplace=True)
df
```

```python
# Vérifier à nouveau les valeurs nulles dans le DataFrame
null_values = df.isnull().sum()

# Afficher le nombre de valeurs nulles pour chaque colonne
print(null_values)
```

```python
df["target_tmp"] = df.shift(-1)["Temp_C"]
```

```python
df = df.ffill()
```

## PREDICTION TEMPERATURE 
### linear regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# Charger les données
X_train = df.drop(['Date/Time','target_vent','target_hum','target_tmp'], axis=1).iloc[:7991]  # Premières 8771 lignes pour l'entraînement
y_train = df['target_tmp'].iloc[:7991]

X_test = df.drop(['Date/Time','target_vent','target_hum','target_tmp'], axis=1).iloc[-768:]  # Dernières 12 lignes pour le test
y_test = df['target_tmp'].iloc[-768:]
# Choisissez le modèle de régression (dans cet exemple, une régression linéaire)
model = LinearRegression()

# Entraînez le modèle sur les données d'entraînement
model.fit(X_train, y_train)

# Faites des prédictions sur les données de test
predictions = model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
rmse = mse ** 0.5  # Root Mean Squared Error
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
print('Mean Absolute Error:', mae)
print('R² Score:', r2)
```

 Voici une analyse des résultats que on a  obtenu:

Mean Squared Error (MSE): 0.4621287260885256

Cette valeur indique la moyenne des carrés des erreurs entre les prédictions et les valeurs réelles. Une MSE plus faible est généralement meilleure, mais pour savoir si 0.462 est bon, il faut comparer avec une baseline ou un modèle simple, ou connaître l'échelle de votre variable cible. Si la valeur de y varie de 0 à 1, par exemple, cela pourrait être assez bon.

Root Mean Squared Error (RMSE): 0.679800504625089

La RMSE est la racine carrée de la MSE et se mesure dans la même unité que la variable cible. Comme pour la MSE, une RMSE plus faible est meilleure. À nouveau, la qualité de cette valeur dépend de l'échelle de y. Si y est dans une plage étroite, un RMSE de 0.68 pourrait être acceptable.

Mean Absolute Error (MAE): 0.48899725640107844

Le MAE représente la moyenne des erreurs absolues. C'est souvent plus intuitif à interpréter que la MSE ou la RMSE. Comme les autres erreurs, une valeur plus faible est souhaitable. La qualité de ce score dépend aussi de l'échelle de votre variable cible.

R² Score: 0.9857230498323407

L'R² score, ou coefficient de détermination, indique la proportion de la variance de y qui est expliquée par le modèle. Une valeur de 0.9857 est très élevée, suggérant que le modèle explique 98.57% de la variance des données. Cela indique une très bonne performance du modèle.
